```python
def similar_words(text,
                    word,
                    window = 8,
                    probability = True,
                    tf_idf=False,
                    dot_product=False,
                    cosine=True):

    vocabulary = sorted(list(set(text)))
    contexts = {}

    for w in vocabulary:
        contexts = []
        for i in range(len(text)):
            if text[i] == w:
                for j in range(i-int(window/2), i):
                    if j >= 0:
                        contexts.append(text[j])
                try:
                    for j in range(i+1, i+int(window/2)):
                        contexts.append(text[j])
                except IndexError:
                    pass
        contexts[w] = contexts

    import numpy as np

    if probability:

        vectors = {}

        for v in vocabulary:
            context = contexts[v]
            vector = []

            for voc in vocabulary:
                vector.append(context.count(voc))
                vector = np.array(vector)
                s = sum(vector)
                vector = vector/s
                vectors[v] = vector

    if dot_product:
        similarities = {}
        v = vectors[word]

        for w in vectors.keys():
            similarities[w] = np.dot(vectors[w], v)
        similarities = (sorted(similarities.items(), key=lambda x: x[1], reverse=True))

        with open('Similar_words_with_dot_product_of_prob_vectors_to_'+word+'.txt', 
        'w', encoding= 'UTF-8') as f:
            for item in similarities:
                f.write(str(item)+'\n')
    
    if cosine:
        similarities = {}
        v1 = vectors[word]

        for w in vectors.keys():
            v2 = vectors[w]
            similarities[w] = np.dot(v1, v2)/(np.linalg.norm(v1)*np.linalg.norm(v2))
        similarities = (sorted(similarities.items(), key=lambda x: x[1], reverse=True))

        with open('Similar_words_with_cosine_of_prob_vectors_to_'+word+'.txt', 
        'w', encoding= 'UTF-8') as f:
            for item in similarities:
                f.write(str(item)+'\n')
    
    if tf_idf:
        import math
        vectors = {}
        for v in vocabulary:
            context = contexts[v]
            vector = []

            for voc in vocabulary:
                tf = context.count(voc)
                idf = math.log(len(text)/len(contexts[voc]))
                vector.append(tf*idf)
                vector = np.array(vector)
                s = sum(vector)
                vector = vector/s
                vectors[v] = vector

        similarities = {}
        v = vectors[word]

        for w in vectors.keys():
            similarities[w] = np.dot(vectors[w], v)
        similarities = (sorted(similarities.items(), key=lambda x: x[1], reverse=True))

        with open('Similar_words_with_dot_product_of_tf_idf_vectors_to_'+word+'.txt',
         'w', encoding= 'UTF-8') as f:
            for item in similarities:
                f.write(str(item)+'\n')

    return similarities

if __name__ == '__main__':
    from normalize import normalize
    from lemmatize import lemmatize

    corpus_path = 'corpus.txt'
    m = 10
    text = normalize(corpus_path, m)
    text = lemmatize(text)

    window = 8
    word = 'gobierno'
    similar_words(text, word, window, probability=True, tf_idf=False,
    dot_product=False, cosine=True)

```
